```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv.imread("../src/butterfly.jpg")
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title('Imagem Original')
plt.axis('off')
plt.show()
```

# Questão 1

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

sift = cv.SIFT_create()
kp = sift.detect(gray, None)

img_out = cv.drawKeypoints(gray, kp, None, flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

plt.imshow(img_out)
plt.title('Sift')
plt.axis('off')
plt.show()
```

# Questão 2

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

fast = cv.FastFeatureDetector_create()
fast.setNonmaxSuppression(False)
kp = fast.detect(gray, None)

img_out = cv.drawKeypoints(gray, kp, None, flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

plt.imshow(img_out)
plt.title('Fast (sem supressão de não-máximos)')
plt.axis('off')
plt.show()
```

# Questão 3

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

fast = cv.FastFeatureDetector_create()
fast.setNonmaxSuppression(True)
kp = fast.detect(gray, None)

img_out = cv.drawKeypoints(gray, kp, None, flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

plt.imshow(img_out)
plt.title('Fast (com supressão de não-máximos)')
plt.axis('off')
plt.show()
```

# Questão 4

```python
brief = cv.xfeatures2d.BriefDescriptorExtractor_create()
kp, des = brief.compute(gray, kp)

print("Primeiro descritor:\n", des[0])
print("\nSegundo descritor:\n", des[1])

# Imprime a quantidade total de descritores
print("\nQuantidade total de descritores:", len(des))
```

# Questão 5

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

orb = cv.ORB_create()
kp = orb.detect(gray, None)
kp, des = orb.compute(gray, kp)

print("Primeiro descritor:\n", des[0])
print("\nSegundo descritor:\n", des[1])

img_out = cv.drawKeypoints(gray, kp, None, flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

plt.imshow(img_out)
plt.title('Orb')
plt.axis('off')
plt.show()
```
